#### **Homework 01:**

##### **Problem 01:**
Calculate the sum and mean of the 3rd row of the matrix.

```python
# Write your code here 

thirdRow = matrix[2, :]  

ThirdRow_sum = np.sum(thirdRow)
ThirdRow_mean = np.mean(thirdRow)

print("Sum of the third row:", ThirdRow_sum)
print("Mean of the third row:", ThirdRow_mean)
```

##### **Problem 02:**
Find the sum and mean of all the rows of the matrix.

```python
# Write your code here

numOfrows = matrix.shape[0]  

for i in range(numOfrows):
    row = matrix[i, :]
    row_sum = np.sum(row)
    print("Row", i+1, "sum:", row_sum)
```

##### **Problem 04:**
Generate the vector `[25, 16, 9, 4, 1]` and then find the square root of each element of the vector using for loop. **Hint:** Use `np.sqrt()` function to find the square root of a number, don't use any other built-in function.

```python
# Write your code here
vector = np.array([25, 16, 9, 4, 1])  


sqrtVector = np.empty_like(vector)

for i in range(vector.size):
    
    sqrtVector[i] = np.sqrt(vector[i])

print("Square root of the vector:", sqrtVector)
```

##### **Problem 05:**
Write a function that takes a number and returns a vector like the previous problem. For example, if the input is 6, then the output should be `[36, 25, 16, 9, 4, 1]`. If the input is 10, then the output should be `[100, 81, 64, 49, 36, 25, 16, 9, 4, 1]`.

```python
# Write your code here
def generate_square_vector(n):
    vector = np.flip(np.arange(1, n+1)**2)
    return vector

# Example usage
number = 10
output = generate_square_vector(number)
print("Output:", output)
```

##### **Problem 08:**
Plot $x + y + z = 1$ for $x, y \in [-3, 3]$ with interval 0.5. Use meshgrid to generate the grid. You can use any built-in function to plot the surface.

```python
# Write your code here

x = np.arange(-3, 3.5, 0.5)
y = np.arange(-3, 3.5, 0.5)

X, Y = np.meshgrid(x, y)

Z = 1 - X - Y

fig = plt.figure()
axis = fig.add_subplot(111, projection='3d')
axis.plot_surface(X, Y, Z)

axis.set_xlabel('x')
axis.set_ylabel('y')
axis.set_zlabel('z')
plt.title('Surface plot of x + y + z = 1')

plt.show()
```

##### **Problem 09:**
Download the image of Lena Forsen from the following link: [https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png](https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png). Then, do the following tasks:

*  Load the image to a numpy array.
*  Find the size of the image.
*  Find the minimum and maximum value of the image.
*  Find the mean and standard deviation of the image.
*  Plot the image using `plt.imshow()` or `display()`.
*  Resize the image to 50 x 50.
*  Convert the image to grayscale.
*  Plot the grayscale image using `plt.imshow()` or `display()`.
*  Save the grayscale image using `imsave()`.
*  Stack the columns of the image to form a vector of size 2500 x 1.

```python
# Write your code here
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color

# Load the image to a numpy array
image_url = "https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png"
image = io.imread(image_url)

# Find the size of the image
image_size = image.shape
print("Image size:", image_size)

# Find the minimum and maximum value of the image
image_min = np.min(image)
image_max = np.max(image)
print("Image minimum value:", image_min)
print("Image maximum value:", image_max)

# Find the mean and standard deviation of the image
image_mean = np.mean(image)
image_std = np.std(image)
print("Image mean value:", image_mean)
print("Image standard deviation:", image_std)

# Plot the image
plt.imshow(image)
plt.axis('off')
plt.title('Original Image')
plt.show()

# Resize the image to 50 x 50
resized_image = io.resize(image, (50, 50), anti_aliasing=True)

# Convert the image to grayscale
grayscale_image = color.rgb2gray(resized_image)

# Plot the grayscale image
plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')
plt.title('Grayscale Image')
plt.show()

# Save the grayscale image
io.imsave('grayscale_image.png', grayscale_image)

# Stack the columns of the image to form a vector
image_vector = image.reshape(-1, 1)
print("Image vector shape:", image_vector.shape)
```

##### **Problem 10:**
Compute 2nd order Taylor approximation of $e^{2x}$ and $e^{3x}$, and $e^{5x}$ around $a = 0$ for $x \in [-1, 1]$ with interval 0.01. Then, determine which of the following is more accurate:
> 1. Calculating the Taylor approximation separately for each $e^{2x}$, $e^{3x}$ then multiplying them together.
> 2. Calculating the Taylor approximation for $e^{5x}$.

Use mean square error (MSE) to determine the accuracy. Also, plot the Taylor approximation for 1st and 2nd method and the original function in the same plot for visual comparison.

```python
# Write your code here
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-1, 1.01, 0.01)

approximation1 = 1 + 2*x + (2*x)**2/2
approximation2 = 1 + 3*x + (3*x)**2/2

approximation3 = 1 + 5*x + (5*x)**2/2

original1 = np.exp(2*x)
original2 = np.exp(3*x)
original3 = np.exp(5*x)

mse1 = np.mean((approximation1 * approximation2 - original1 * original2)**2)
mse2 = np.mean((approximation3 - original3)**2)

print("MSE for method 1:", mse1)
print("MSE for method 2:", mse2)

plt.figure(figsize=(10, 6))

plt.plot(x, approximation1 * approximation2, label='Taylor Approximation (Method 1)')

plt.plot(x, approximation3, label='Taylor Approximation (Method 2)')

plt.plot(x, original1 * original2, label='Original Function (Method 1)', linestyle='dashed')
plt.plot(x, original3, label='Original Function (Method 2)', linestyle='dashed')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Taylor Approximation vs Original Function')
plt.legend()
plt.grid(True)
plt.show()
```
